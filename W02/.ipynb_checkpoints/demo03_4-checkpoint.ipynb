```python
# 先裝 mysql的connection driver
# pip install PyMySQL

from sqlalchemy import create_engine

my_db = create_engine(mysql_db_url)
```

### Create
- 先思考要有什麼欄位?
- 我們有 ['dtime', 'v1', 'v2', 'v3', 'a1', 'a2', 'a3', 'total_e']
- 他們是什麼格式的資料

```python
# 我們把tsp, pm10, lead都記下來
table_name = 'malo_khpower_1209'
sql_cmd_str = "CREATE TABLE IF NOT EXISTS %s(dtime varchar(20), v1 float, v2 float, v3 float, a1 float, a2 float, a3 float, total_e float, PRIMARY KEY(dtime))" %(table_name)
resultProxy = my_db.execute(sql_cmd_str)
resultProxy.__dict__
```

```python
# 做upsert操作:
for item in data_list[1:]:
    dtime = item[0]
    v1 = item[1]
    v2 = item[2]
    v3 = item[3]
    a1 = item[4]
    a2 = item[5]
    a3 = item[6]
    total_e = item[7]
    sql_cmd_str = "insert into %s (dtime, v1, v2, v3, a1, a2, a3, total_e) values('%s', '%s', '%s', '%s', '%s', '%s', '%s', '%s')" %(
        table_name, dtime, v1, v2, v3, a1, a2, a3, total_e)
    sql_cmd_str = sql_cmd_str + (" ON DUPLICATE KEY UPDATE v1='%s', v2='%s', v3='%s', a1='%s', a2='%s', a3='%s', total_e='%s'" %(v1, v2, v3, a1, a2, a3, total_e))
    resultProxy=my_db.execute(sql_cmd_str)
    #print(sql_cmd_str)
    
```

```python
# 做upsert操作 --> 這是之前的demo，還記得在防止什麼事情嗎? 思考一下要怎麼使用在這邊

for item in data_list[1:]:
    sid = item[0]#['Name']
    dtime = item[1]#['Date']
    tsp = item[2]#['TSP']
    if tsp=='':
        tsp = 'NULL'
    pm10 = item[3]#['PM10']
    if pm10=='':
        pm10 = 'NULL'
    lead = item[7]#['Lead']
    if lead=='':
        lead = 'NULL'
    sql_cmd_str = "insert into %s (sid, dtime, tsp, pm10, lead) values('%s', '%s', %s, %s, %s)" %(
        table_name, sid, dtime, tsp, pm10, lead)
    sql_cmd_str = sql_cmd_str + (" ON DUPLICATE KEY UPDATE tsp=%s, pm10=%s, lead=%s" %(tsp, pm10, lead))
    resultProxy=my_db.execute(sql_cmd_str)
    print(sql_cmd_str)
    
```

```python
# 找出某個時間區間的資料: where
sql_cmd_str = "select * from %s where dtime<'2022-12-01 06:00:00'" %(table_name)
resultProxy=my_db.execute(sql_cmd_str)
data = resultProxy.fetchall()
print('-- data --')
#print(data)
for item in data:
    print(item)
```

### 把趨勢圖畫出來，怎麼圖很怪?

```python
import matplotlib.pyplot as plt

total_e_list = []
dtime_list = []
for item in data:
    total_e_list.append(item['total_e'])
    dtime_list.append(item['dtime'])
    
plt.plot(dtime_list, total_e_list)
plt.xlabel('month')
plt.ylabel('total_energy')
plt.xticks(dtime_list, rotation=90)
plt.grid()
```

### 畫三相電壓的趨勢

```python
import matplotlib.pyplot as plt

total_e_list = []
v1_list = []
v2_list = []
v3_list = []
dtime_list = []
for item in data:
    total_e_list.append(item['total_e'])
    v1_list.append(item['v1'])
    v2_list.append(item['v2'])
    v3_list.append(item['v3'])
    dtime_list.append(item['dtime'])
    
plt.plot(dtime_list, v1_list)
plt.plot(dtime_list, v2_list)
plt.plot(dtime_list, v3_list)
plt.xlabel('time')
plt.ylabel('voltage')
plt.xticks(dtime_list[::6], rotation=90)
plt.grid()
```

### 太小? 利用 `plt.subplots(figsize=(12,8))` 把圖變大

```python
import matplotlib.pyplot as plt

total_e_list = []
v1_list = []
v2_list = []
v3_list = []
dtime_list = []
for item in data:
    total_e_list.append(item['total_e'])
    v1_list.append(item['v1'])
    v2_list.append(item['v2'])
    v3_list.append(item['v3'])
    dtime_list.append(item['dtime'])

plt.subplots(figsize=(12,8))
plt.plot(dtime_list, v1_list, 'r')
plt.plot(dtime_list, v2_list, 'g')
plt.plot(dtime_list, v3_list, 'b')
plt.xlabel('time')
plt.ylabel('voltage')
plt.xticks(dtime_list[::6], rotation=90)
plt.grid()
plt.show()
```

### 如果這張圖需要放到簡報中，我們可能需要加上圖例
- legend() : 自動畫出圖例
- label : 每一條線的名稱
- 上面的圖有點過大，再縮小一點(figsize的default: [6.4, 4.8]，我們改成[8, 6])

```python
import matplotlib.pyplot as plt

total_e_list = []
v1_list = []
v2_list = []
v3_list = []
dtime_list = []
for item in data:
    total_e_list.append(item['total_e'])
    v1_list.append(item['v1'])
    v2_list.append(item['v2'])
    v3_list.append(item['v3'])
    dtime_list.append(item['dtime'])

plt.subplots(figsize=(8,6))
plt.plot(dtime_list, v1_list, 'r', label='v1')
plt.plot(dtime_list, v2_list, 'g', label='v2')
plt.plot(dtime_list, v3_list, 'b', label='v3')
plt.xlabel('time')
plt.ylabel('voltage')
plt.xticks(dtime_list[::6], rotation=90)
plt.grid()
plt.legend()
plt.show()
```
