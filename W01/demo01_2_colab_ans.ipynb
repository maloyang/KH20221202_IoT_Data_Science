```python
import folium
import geocoder
```

```python
# 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('新竹市').latlng
```

```python
print(location)
```

```python
m = folium.Map(location=location, zoom_start=15)
```

```python
for item in data_list[1:]:
    try:
        # item[2]=緯度, item[1]=經度, item[0]=名稱
        folium.Marker([float(item[2]), float(item[1])], popup=item[0]).add_to(m)
    except Exception as e:
        print(e.args)    
```

- 練習題: 把地址訊息加入地圖中

```python
import folium
import geocoder

# 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('新竹市').latlng

m = folium.Map(location=location, zoom_start=15)

for item in data_list[1:]:
    try:
        # item[2]=緯度, item[1]=經度, item[0]=名稱, item[4]=地址
        folium.Marker([float(item[2]), float(item[1])], popup=(item[0]), tooltip=item[4] ).add_to(m)
    except Exception as e:
        print(e.args)
```
