```python
import csv

fn = 'bike_02_新北市.csv.2'
with open(fn, 'r', encoding='utf8') as csvfile:
    rows = csv.reader(csvfile)
    data_list = list(rows)
```

```python
# 請在這邊填入你的程式

import folium
import geocoder

location = geocoder.osm('新北市').latlng
m = folium.Map(location=location, zoom_start=13)

for item in data_list[1:]:
    try:
        """
        ['1002',          0
        '汐止火車站',       1
        '44',           2
        '28',           3
        '汐止區',           4
        '20221119114321',       5
        '25.068914',        6
        '121.662748',       7
        '南昌街/新昌路口(西側廣場)', 8
        'Xizhi Dist.',          9
        'Xizhi Railway Station',    10
        'Nanchang St./Xinchang Rd.',    11
        '16',               12
        '1']                13
        """
        # item[2]=緯度, item[1]=經度, item[0]=名稱
        
        folium.Marker( [float(item[6]), float(item[7])], 
                tooltip=f"({item[1]}) {item[3]}/{item[2]}",
                icon=folium.Icon(color='green', prefix='fa', icon='fa-bicycle')
            ).add_to(m)
        
    except Exception as e:
        print(e.args)    
```

```python
# 把程式重整!!

# 請在這邊填入你的程式

import folium
import geocoder

location = geocoder.osm('新北市').latlng
m = folium.Map(location=location, zoom_start=13)

for item in data_list[1:]:
    try:
        lat = float(item[6])
        lng = float(item[7])
        name = item[1]
        total = item[2]
        n = item[3]
        msg = f"({name}) {n}/{total}"
        folium.Marker( [lat, lng], 
                tooltip=msg,
                icon=folium.Icon(color='green', prefix='fa', icon='fa-bicycle')
            ).add_to(m)
        
    except Exception as e:
        print(e.args)    
```
